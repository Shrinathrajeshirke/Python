```python
### importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
## reading dataset

df = pd.read_csv('D:/KNMaterial/my_python/4-ML_Models/datasets/Algerian_forest.csv',header=1)

df.head()
```

```python
## basic info of dataset

df.info()
```

### Data cleaning

```python
## missing values

df.isnull().sum()
```

The dataset is converted into two sets based on region from 122th index, we can make a new column based on the region

1: "Bejala region"

2: "Sidi-Bel Abbes region"


```python
df.loc[:122, "Region"]=0
df.loc[122:, "Region"]=1
df[['Region']] = df[['Region']].astype(int)
df.info()
```

```python
### missing values

df[df.isnull().any(axis=1)]
```

```python
## drop null values and reset index
df=df.dropna().reset_index(drop=True)

## check null values
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)

df.iloc[[122]]
```

```python
## columns fix spaces

print('columns before update: \n', df.columns)

df.columns= df.columns.str.strip()

print('columns after update: \n', df.columns)
```

```python
## change data types

df.head(2)
```

```python
df.columns
```

```python
## change it to int and float type

df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
df[['Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI']]=df[['Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI']].astype(float)

df.info()
```

```python
## descriptive stats

df.describe()
```

```python
## save the dataset

df.to_csv('D:/KNMaterial/my_python/4-ML_Models/datasets/Algerian_forest_cleaned.csv',index=False)
```

### Exploratory Data Analysis

```python
## drop columns

df_copy = df.drop(columns=['day','month','year'],axis=1)
df_copy.head(2)
```

```python
## encoding of the categories in classes


df_copy['Classes']=np.where(df_copy['Classes'].str.contains('not fire'),0,1)
df_copy['Classes'].value_counts()
```

```python
## plot density plot for all features

plt.style.use('seaborn-v0_8-darkgrid')
df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
## percentage of pie chart

percentage= df_copy['Classes'].value_counts(normalize=True)*100
```

```python
## plot pie chart

classlabels = ['Fire','Not Fire']
plt.Figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title('Percentage Pie chart ')
plt.show()
```

```python
## correlation 

df_copy.corr()
```

```python
sns.heatmap(df_copy.corr())
```

```python
### boxplots

sns.boxplot(df_copy['FWI'])
```
