```python
### importing libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
%matplotlib inline
```

```python
## read dataset

df = pd.read_excel('datasets/flight_price.xlsx')

## top 5 records
df.head()
```

```python
## summary of dataset
df.info()
```

```python
## descriptive stats
df.describe()
```

```python
## feature engineering

#extracting date, month and year 
#creating new columns
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]

#type casting to int for all 
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)

#display df info
df.info()
```

```python
## drop date of journey

df.drop("Date_of_Journey", axis=1, inplace=True)
df.head(2)
```

```python
##extracting hour and minutes

## remove extra things from arrival time 
#to keep only hour and time

df['Arrival_Time']= df['Arrival_Time'].apply(lambda x:x.split(' ')[0])

df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min'] = df['Arrival_Time'].str.split(':').str[1]

## type casting to int
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_min']=df['Arrival_min'].astype(int)

##  drop arrival time
df.drop('Arrival_Time',axis=1, inplace=True)

df.head(2)
```

```python
##extracting hour and minutes

## remove extra things from departure time 
#to keep only hour and time

df['Dep_Time']= df['Dep_Time'].apply(lambda x:x.split(' ')[0])

df['Dep_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_min'] = df['Dep_Time'].str.split(':').str[1]

## type casting to int
df['Dep_hour']=df['Dep_hour'].astype(int)
df['Dep_min']=df['Dep_min'].astype(int)

##  drop Departute time
df.drop('Dep_Time',axis=1, inplace=True)

df.head(2)
```

```python
## Total stops

#unique total stops
df['Total_Stops'].unique()
```

```python
#missing values

df['Total_Stops'].isnull().sum()
```

```python
## replace values with number of stops

df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,np.nan:1})

##check missing values
df[df['Total_Stops'].isnull()]
```

```python
## dropping Route

df.drop('Route',axis=1, inplace=True)
df.head(2)
```

```python
## Duration

df['hr'] = df['Duration'].str.extract(r'(\d+)h').fillna(0).astype(int)
df['min'] =df['Duration'].str.extract(r'(\d+)m').fillna(0).astype(int)

df['hr'] = df['hr'].apply(lambda x:x*60)
df['Total_Duration'] = df['hr']+df['min']
df.head(2)
```

```python
## dropping duration column
df.drop('Duration', axis=1,inplace=True)
df.info()
```

```python
## Airline, Source, additional_info

print(df['Airline'].unique())
print(df['Additional_Info'].unique())
print(df['Source'].unique())
```

```python
## onehot encoding

from sklearn.preprocessing import OneHotEncoder

#instance of encoder
encoder = OneHotEncoder()

##encoding
encoded_df = pd.DataFrame(encoder.fit_transform(df[['Airline','Source','Destination']]).toarray(), columns=encoder.get_feature_names_out())
df = pd.concat([encoded_df, df],axis=1)
df.info()
```

```python
## dropping columns 
df.drop(['Airline','Source','Destination'], axis=1,inplace=True)
df.info()
```
