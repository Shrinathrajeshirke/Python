### Stemming

Stemming is an important text-processing technique that reduces words to their base or root form by removing prefixes and suffixes.

```python
### Porter stemmer

words = ["running", "jumps", "happily", "programming", "fairly"]

from nltk.stem import PorterStemmer
porter = PorterStemmer()

for word in words:
    print(word+"----->"+porter.stem(word))
```

Major disadvantage -- it is not creating meaningful words in some cases

```python
### RegexpStemmer

from nltk.stem import RegexpStemmer
regex = RegexpStemmer('ing$|s$|e$|able$', min=4)
for word in words:
    print(word+"----->"+regex.stem(word))
```

```python
### snowball stemmer

from nltk.stem import SnowballStemmer
snow = SnowballStemmer('english')
for word in words:
    print(word+"----->"+snow.stem(word))
```
